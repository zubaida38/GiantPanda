```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D'],
    'math': [60,89,82,70],
    'physics': [66,95,83,66],
    'chemistry': [61,91,77,70]
})
df1
```

```python
df2 = pd.DataFrame({
    'name': ['E', 'F', 'G', 'H'],
    'math': [66,95,83,66],
    'physics': [60,89,82,70],
    'chemistry': [90,81,78,90]
})
df2
```

## 1. Dealing with index and axis

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1, df2],ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

## 2. Avoiding duplicate indices

```python
try:
    pd.concat([df1,df2], verify_integrity=True)
except ValueError as e:
    print('ValueError:', e)
```

## 3. Adding a hierarchical index with keys and names options

```python
res = pd.concat([df1,df2],keys=['Year 1','Year 2'])
res
```

```python
res.loc['Year 1']
```

```python
pd.concat(
    [df1, df2], 
    keys=['Year 1', 'Year 2'],
    names=['Class', None],
)
```

```python
pd.concat(
    [df1, df2], 
    keys=['Year 1', 'Year 2'],
    names=['Class', None],
).reset_index(level=0)
```

```python
# Pass a string to level
pd.concat(
    [df1, df2], 
    keys=['Year 1', 'Year 2'],
    names=['Class', None],
).reset_index(level='Class')
```

## 4. Columns matching and sorting

```python
df1 = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D'],
    'chemistry': [61,91,77,70],
    'physics': [66,95,83,66],
    'math': [60,89,82,70],
})
df1
```

```python
df2 = pd.DataFrame({
    'name': ['E', 'F', 'G', 'H'],
    'math': [66,95,83,66],
    'physics': [60,89,82,70],
    'chemistry': [90,81,78,90]
})
df2
```

```python
pd.concat([df1, df2])
```

```python
# sort by columns name
pd.concat([df1, df2], sort=True)
```

```python
# custom sort
custom_sort = ['math', 'chemistry', 'physics', 'name']
res = pd.concat([df1, df2])
res[custom_sort]
```
