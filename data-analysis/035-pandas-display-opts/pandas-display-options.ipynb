## 1. Showing more rows

```python
df
```

```python
# By default, Pandas doesn't want to over run your display
# It displays the first 5 rows and last 5 rows elipiss 
pd.get_option('display.max_rows')
```

```python
# Use attribute
pd.options.display.max_rows
```

```python
# Set to a value, for example 200 rows
pd.set_option('display.max_rows', 200)

# pd.options.display.max_rows = 200
```

```python
# To show all rows 
pd.set_option('display.max_rows', None)

# Use attribute
# pd.options.display.max_rows = None
```

```python
# To reset
pd.reset_option('display.max_rows')
```

## 5. Formatting the display for numbers

### Formatting big number with comma

```python
df_test = pd.DataFrame([100000.12345, 2000000.5678, 34500000.6789, 456000.7890],
                  index=[1,2,3,4],
                  columns=['test'])
df_test
```

```python
pd.set_option('display.float_format',  '{:,}'.format)
df_test
```

### Setting decimal format

```python
# note value are rounded
pd.set_option('display.float_format',  '{:,.2f}'.format)
df_test
```

### Formatting with a dollar or percent sign

```python
pd.reset_option('display.float_format')

df_test = pd.DataFrame([12.3456, 234.5678, 345.6789, 456.7890],
                  index=[1,2,3,4],
                  columns=['test'])
df_test
```

```python
pd.set_option('display.float_format', '{:.2f}%'.format)
df_test
```

```python
pd.set_option('display.float_format', '${:.2f}'.format)
df_test
```

```python
pd.reset_option('display.float_format')
```
